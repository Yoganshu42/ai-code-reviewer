```python
test_code = ("def find_max(nums):"
                "max_val = None"
                "for n in nums:"
                    "if max_val is None or n > max_val:"
                        "max_val = n"
                "return max_val")

prompt = (
     f"{test_code}\n\n"
    "You are a code reviewer. Read the code change below and write one concise review comment in plain english text.\n\n"
    "suggestion:\n\n"
    
)

inputs = tokenizer(prompt, return_tensors="pt", truncation=True)

out = model.generate(
    **inputs,
    max_new_tokens=60,
    num_beams=1,
    do_sample=False,
    early_stopping=True,
)

decoded = tokenizer.decode(out[0], skip_special_tokens=True)

# extract only the generated part
comment = decoded.split("Comment:")[-1].strip()
print("SUGGESTION:", comment)
```
